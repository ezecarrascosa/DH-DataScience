```python
# 8.1 Analizamos varios ejemplos. Determinamos reglas que aplicaremos mas adelante
display(df_temp_split.sample(10))
display(df_temp_split.sample(10))
display(df_temp_split.sample(10))
# Determinamos que podemos tener las siguiente columnas 
# Pais - Localidad/Provincia(Arg) - Partido - Barrio - Barrio Cerrado -Zona(CABA)
```

```python
# 9 - Cambiamos los nombres de las columnas
df_temp_split=df_temp_split.rename(columns={1: 'Pais',2: 'Localidad/Provincia(Arg)',3:'Partido',4: 'Barrio', 5:'Barrio Cerrado'})
```

```python
# 10 Agregamos una nueva columna Zona(CABA). ya que vimos que por ej Palermo tiene divisiones ej Palermos Soho 
df_temp_split['Zona(CABA)']=''
df_temp_split.head(2)
```

```python
# 11 Creamos varias condiciones que vamos a utilizar para formatear las columnas
cond_localidad_capital = df_temp_split['Localidad/Provincia(Arg)']=='capital federal'
cond_localidad_nocapital = df_temp_split['Localidad/Provincia(Arg)']!='capital federal'
cond_barrio_vacio = df_temp_split['Barrio']==''
cond_barrio_novacio = df_temp_split['Barrio']!=''
cond_partido_novacio = df_temp_split['Partido']!=''
cond_barrio_equal_zona = df_temp_split['Zona(CABA)']==df_temp_split['Barrio']
```

```python
# 12 Aplicamos las reglas

# Regla 1 : es capital, tenemos y barrio y tenemos partido -> el partido es la zona : ejemplo Palermo - Palermo Soho
df_temp_split.loc[cond_localidad_capital & cond_partido_novacio & cond_barrio_novacio,'Zona(CABA)']=df_temp_split.loc[cond_localidad_capital & cond_partido_novacio & cond_barrio_novacio,'Partido']

# Regla 2 : es capital, no tenemos barrio y tenemos partido -> el partido es el barrio
df_temp_split.loc[cond_localidad_capital & cond_partido_novacio & cond_barrio_vacio,'Barrio']=df_temp_split.loc[cond_localidad_capital & cond_partido_novacio & cond_barrio_vacio,'Partido']

# Regla 3 : es capital, y tenemos partido -> lo limpiamos porque con las reglas anteriores ya nos encargamos del partido
df_temp_split.loc[cond_localidad_capital & cond_partido_novacio,'Partido']=''

# Regla 4 : es capital, y la zona = al barrio, limiamos la zona y dejamos solo el barrio
df_temp_split.loc[cond_localidad_capital & cond_barrio_equal_zona,'Zona(CABA)']=''

# Analisis mas profundo de gran buenos aires y Provincias de Argentina --> PENDIENTE
```

```python
# 13 - Acomodamos las columnas 
df_temp_split = df_temp_split[['Pais','Localidad/Provincia(Arg)','Partido','Zona(CABA)','Barrio','Barrio Cerrado']]

df_temp_split.head(10)
```

```python
# 15 limpiamos los valores nulos
indices = pd.isnull(df_temp_split["Pais"])
df_temp_split.at[indices,"Pais"] = ''

indices = pd.isnull(df_temp_split["Zona(CABA)"])
df_temp_split.at[indices,"Zona(CABA)"] = ''

indices = pd.isnull(df_temp_split["Barrio"])
df_temp_split.at[indices,"Barrio"] = ''

indices = pd.isnull(df_temp_split["Barrio Cerrado"])
df_temp_split.at[indices,"Barrio Cerrado"] = ''

indices = pd.isnull(df_temp_split["Localidad/Provincia(Arg)"])
df_temp_split.at[indices,"Localidad/Provincia(Arg)"] = ''

display(df_temp_split.sample(10))
display(df_temp_split.sample(10))
```

```python
# 2.9 convertimos los pesos a dolares

# Colocamos la cotiz que vamos a usar. esto hay que cambiarlo a dinamico. Si estimamos -por ej- cotiz a julio puede ser 60 
# Tengamos en cuenta que la cotiz que uso para pasar de pesos a dolares fue de 17,644
cotiz_dolar=45

# Tomamos solo lo que esta en pesos
x=df_temp_precios.loc[cond_ars,:]

# convertimos price_aprox_local_currency a numerico y dividimos x la cotiz del dolar
df_temp_precios.loc[cond_ars,"price_aprox_usd"]=pd.to_numeric(x["price_aprox_local_currency"]) / cotiz_dolar
```

```python
# 2.10 Chequeamos los valores
df_temp_precios.loc[cond_ars,:].head(19)
```

```python
# 2.11 necesitamos el valor del metro cuadrado. por sabemos que el mejor valor que tenemos es price_aprox_usd

#cond_price = pd.to_numeric(df_temp_precios["price"]) > 0
cond_price = pd.to_numeric(df_temp_precios["price_aprox_usd"]) > 0
cond_m2    = pd.to_numeric(df_temp_precios["surface_total_in_m2"]) > 0


df_temp_precios.loc[cond_price & cond_m2,"M2 USD"]=df_temp_precios.loc[cond_price & cond_m2,"price_aprox_usd"]/df_temp_precios.loc[cond_price & cond_m2,"surface_total_in_m2"]
df_temp_precios.head(5)
```

```python
# 2.12 Obtenemos superficie descubierta
cond_sup_total      = pd.to_numeric(df_temp_precios["surface_total_in_m2"]) > 0
cond_sup_covered    = pd.to_numeric(df_temp_precios["surface_covered_in_m2"]) > 0

df_temp_precios.loc[cond_sup_total & cond_sup_covered,"Superficie descubierta"]=df_temp_precios.loc[cond_sup_total & cond_sup_covered,"surface_total_in_m2"] - df_temp_precios.loc[cond_price & cond_m2,"surface_covered_in_m2"]

df_temp_precios.head(3)
```

```python
# 2.13 renombramos columnas
df_temp_precios=df_temp_precios.rename(columns={"price": 'Precio',"currency":'Moneda',"price_aprox_local_currency":"Price aprox ARS","price_aprox_usd":"Price aprox USD","surface_total_in_m2": 'Sup Total','surface_covered_in_m2':'Sup Cubierta','Superficie descubierta':'Sup descubierta'})
df_temp_precios.head(10)
```

```python
# 2.14 borramos columnas que no nos interesan
del df_temp_precios['price_usd_per_m2']
del df_temp_precios['price_per_m2']
```

```python
# 2.15 Acomodamos las columnas

df_temp_precios = df_temp_precios[['Precio','Moneda','Price aprox ARS','Price aprox USD','Sup Total','Sup Cubierta','Sup descubierta','M2 USD','Opc.pago','Requiere Analisis']]
```
